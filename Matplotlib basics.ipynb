```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic graph

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]

#Resize your graph dpi is pixels per inc
plt.figure(figsize=(5,3),dpi=150)

#plt.plot(x,y, label ='2x',color ='red',linestyle='--' ,linewidth = 2,marker='*',markersize=10, markeredgecolor='blue')

#use shorthand notation
#fmt ='[color][marker][line]'
plt.plot(x,y,'r^--',label='2x')
#plt.show() (separate plt.show() will give separate graphs, single will put multiple lines into one graph)
## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[5:], x2[5:]**2,'b',label='x^2')
plt.plot(x2[:6],x2[:6]**2,'y--')


plt.title('Our first graph', fontdict= {'fontname': 'Comic Sans MS','fontsize': 10})
plt.xlabel('X axis', fontdict= {'fontname': 'Comic Sans MS','fontsize': 12})
plt.ylabel('Y axis',fontdict= {'fontname': 'Comic Sans MS','fontsize': 12})


#customize x ticks and y ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png',dpi=300)
plt.show()
```

#### Bar chart

```python
labels =['A','B','C']
values=[1,4,2]
bars= plt.bar(labels, values)

patterns =['/','*','o']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('*')
#bars[2].set_hatch('o')
plt.figure(figsize=(6,4))
plt.show()
```
